```python
def paindrome(input_string):
    input_string=input_string.lower().replace(" ", "").replace("á", "a").replace("é", "e").replace("í", "i").replace("ó", "o").replace("ú", "u")
    if input_string == input_string[::-1]: print(True)
    else: return print(False)

# Pruebas Unitarias
paindrome("Anita lava la tina")  # Debería imprimir                             True
paindrome("A mamá Roma le aviva el amor a mamá y a papá")  # Debería imprimir   True
paindrome("Hola mundo")  # Debería imprimir                                     False
paindrome("Aibofobia")  # Debería imprimir                                      True
paindrome("Amor a Roma")  # Debería imprimir                                    True
paindrome("Esto no es un palíndromo")  # Debería imprimir                       False
paindrome("")  # Debería imprimir                                               True
```
