```python
def Entrada(ruta_archivo):
    with open(ruta_archivo, 'r') as archivo:
        datos_entrada = archivo.read().strip().split('\n')
    return datos_entrada

def Formula_1(datos_entrada):
    indice = 0
    resultados_globales = []
    while indice < len(datos_entrada):
        try:
            num_carreras, num_pilotos = map(int, datos_entrada[indice].split())
            indice += 1
        except ValueError:
            print(f"Formato incorrecto en la línea {indice + 1}: {datos_entrada[indice]}")
            indice += 1
            continue

        if num_carreras == 0 and num_pilotos == 0:
            break

        resultados_carreras = []
        for _ in range(num_carreras):
            try:
                resultados_carreras.append(list(map(int, datos_entrada[indice].split())))
                indice += 1
            except ValueError:
                print(f"Formato incorrecto en la línea {indice + 1}: {datos_entrada[indice]}")
                indice += 1

        if indice >= len(datos_entrada):
            print(f"Se esperaban más líneas para sistemas de puntuación después de la línea {indice}")
            break

        try:
            num_sistemas = int(datos_entrada[indice])
            indice += 1
        except ValueError:
            print(f"Formato incorrecto para el número de sistemas de puntuación en la línea {indice + 1}: {datos_entrada[indice]}")
            indice += 1
            continue

        for _ in range(num_sistemas):
            if indice < len(datos_entrada):
                try:
                    sistema = list(map(int, datos_entrada[indice].split()))
                    limite_puntos = sistema[0]
                    puntos = sistema[1:limite_puntos + 1]
                    indice += 1
                except (ValueError, IndexError):
                    print(f"Formato incorrecto o incompleto en la línea {indice + 1}: {datos_entrada[indice]}")
                    indice += 1
                    continue

                puntajes_pilotos = [0] * num_pilotos
                for carrera in resultados_carreras:
                    for indice_piloto, posicion in enumerate(carrera):
                        if posicion <= limite_puntos:
                            try:
                                puntajes_pilotos[indice_piloto] += puntos[posicion - 1]
                            except IndexError:
                                print(f"Posición {posicion} excede el límite de puntos {limite_puntos}")

                puntaje_maximo = max(puntajes_pilotos)
                ganadores = [str(i + 1) for i, puntaje in enumerate(puntajes_pilotos) if puntaje == puntaje_maximo]
                resultados_globales.append(ganadores)
            else:
                print(f"Se esperaban más líneas para sistemas de puntuación después de la línea {indice}")
                break

    return resultados_globales


#ruta_archivo = '/content/drive/MyDrive/Colab Notebooks/Prueba GBM/Ejercicio 2/datos_entrada.txt'
#ruta_archivo_pruebas = '/content/drive/MyDrive/Colab Notebooks/Prueba GBM/Ejercicio 2/datos_prueba.txt'
#ruta_archivo_pruebas_en_blanco = '/content/drive/MyDrive/Colab Notebooks/Prueba GBM/Ejercicio 2/datos_prueba_en_balnco.txt'

datos_entrada = Entrada(input())
resultados = Formula_1(datos_entrada)
for resultado in resultados:
    print(' '.join(resultado))


```
