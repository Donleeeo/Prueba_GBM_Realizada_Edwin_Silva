```python
def saltos_requeridos(x):
    suma_total = 0
    saltos = 0

    while True:
        saltos = saltos+ 1
        suma_total = suma_total + saltos

        # Comprobar si la suma actual ha alcanzado o posicion x
        if suma_total >= x:

            # Compruebo si la diferencia es par, si es así, puedo parar
            if (suma_total - x) % 2 == 0:
                break

            # Verifico si slo un paso resuelve la diferencia

            elif suma_total - x == 1:
                saltos = saltos+ 1  # Saltos incrementales para tener en cuenta el paso hacia atrás
                break
    return saltos
```

```python
print(f"Saltos requeridos en posicion 1: {saltos_requeridos(1)}")
print(f"Saltos requeridos en posicion 2: {saltos_requeridos(2)}")
print(f"Saltos requeridos en posicion 3: {saltos_requeridos(3)}")
print(f"Saltos requeridos en posicion 4: {saltos_requeridos(4)}")
print(f"Saltos requeridos en posicion 5: {saltos_requeridos(5)}")
print(f"Saltos requeridos en posicion 6: {saltos_requeridos(6)}")
print(f"Saltos requeridos en posicion 7: {saltos_requeridos(7)}")
print(f"Saltos requeridos en posicion 8: {saltos_requeridos(8)}")
print(f"Saltos requeridos en posicion 9: {saltos_requeridos(9)}")
print(f"Saltos requeridos en posicion 10: {saltos_requeridos(10)}")
```
